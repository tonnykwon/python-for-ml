```python
# visualization
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X,Y)
plt.show()
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in Y]

plt.plot(X_1,Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_2, c="b")
ax_2.plot(X_2, Y_2, c="g")
plt.show()
```

```python
# set Color
Y_1 = X_1
Y_2 = [value+100 for value in X_2]

plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")
plt.show()
```

```python
# set linestyle
plt.plot(X_1, Y_1, c="b", linestyle="dashed", label="line 1")
plt.plot(X_2, Y_2, c="r", ls="dotted", label="line 2")
plt.legend(shadow=True, fancybox=False, loc="upper right")

plt.text(50,80, "Line_1")
plt.annotate(
    'Line_2', xy=(50, 150), xytext=(20, 175),
    arrowprops = dict(facecolor='black', shrink=0.05))


plt.title("Two lines")
plt.xlabel("$X_line$")

plt.show()
```

```python
plt.plot(X_1, Y_1, color ="b", linestyle = "dashed", label='line_1')
plt.plot(X_2, Y_2, color ="r", linestyle = "dotted", label='line_2')
plt.legend(shadow=False, fancybox=True, loc="lower right")

plt.grid(True, lw=0.4, ls="--", c=".90")
plt.xlim(-1000, 2000)

plt.show()
```

```python
# Scatter
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1 ], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1 ], c="r", marker="o")
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
# Bar chart
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)
```

```python
plt.bar(X +0.00, data[0], color="b", width=0.5)
plt.bar(X +0.5, data[1], color="g", width=0.5)
plt.bar(X +1, data[2], color="r", width=0.5)
plt.xticks(X+0.5, ["A","B","C","D"])
plt.show()
```

```python
color_list= ["b", "g", "r"]
data_label = ["A", "B","C"]
for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0),
           color=color_list[i], label=data_label[i])
plt.legend()
plt.show()
```
