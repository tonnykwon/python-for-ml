```python
dataurl = "https://www.shanelynn.ie/wp-content/uploads/2015/06/phone_data.csv"
df_phone = pd.read_csv(dataurl)
df_phone.head()
```

```python
import dateutil
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.groupby('month')["duration"].sum()
```

```python
df_phone[df_phone['item'] == 'call'].groupby('month')['duration'].sum() 
```

```python
df_phone.groupby(['month', 'item'])['duration'].sum()
```

```python
df_phone.groupby(['month', 'item'])['date'].count().unstack()
```

```python
df_phone.groupby('month', as_index=False).agg({"duration":"sum"})
```

```python
df_phone.groupby(['month','item']).agg({'duration':sum,
                                       'network_type':"count",
                                       'date': 'first'})
```

```python
df_phone.groupby(['month','item']).agg({"duration": [min],
                                       'network': 'count',
                                       'date': [min, 'first', 'nunique']})
```

```python
grouped = df_phone.groupby("month").agg({"duration": [min, max, np.mean]})
grouped
```

```python
grouped.columns = grouped.columns.droplevel(level=0)
grouped
```

```python
grouped.rename(columns={"min": "min_duration", "max": "max_duration", "mean": "mean_duration"})
```

```python
# pivot table
df_phone.head()
```

```python
df_phone.pivot_table(["duration"],
                    index =[df_phone.month, df_phone.item],
                    columns=df_phone.network, aggfunc="sum",
                    fill_value=0)
```

```python
# crosstab
df_phone.pivot_table(["duration"], index=df_phone.network_type,
                    columns = df_phone.item,
                    aggfunc="sum", fill_value=0)
```

```python
pd.crosstab(index=df_phone.network_type, columns= df_phone.item,
            values = df_phone.duration, aggfunc="first").fillna(0)
```

```python
# merge
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on="subject_id")
```

```python
# when key names are different
pd.merge(df_a, df_b, left_on = 'subject_id', right_on = 'subject_id')
```

```python
pd.merge(df_a, df_b, on="subject_id", how='left')
```

```python
pd.merge(df_a, df_b, on="subject_id", how='inner')
```

```python
pd.merge(df_a, df_b, right_index= True, left_index=True)
```

```python
# concat
df_new = pd.concat([df_a, df_b])
df_new.reset_index()
```
