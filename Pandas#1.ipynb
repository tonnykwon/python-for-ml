```python
dataurl = 'https://raw.githubusercontent.com/rstudio/Intro/master/data/wages.csv'
df = pd.read_csv(dataurl)
```

```python
df.head()
```

```python
df.sex.unique()
```

```python
df["sec_code"] = df.sex.map({"male":0, "female":1})
df.head(5)
```

```python
df.sex.replace({"male":0, "female":1}).head()
```

```python
df.sex.replace(["male", "female"], [0,1], inplace=True)
df
```

```python
del df["sec_code"]
```

```python
df.head()
```

```python
# apply & applymap
df_info = df[["earn", "height", "age"]]
df_info.head()
```

```python
f = lambda x : x.max() - x.min()
df_info.apply(f)
```

```python
df_info.apply(sum)
```

```python
df_info.sum()
```

```python
def f(x):
    return Series([x.min(), x.max(), x.mean()], index =["min", "max", "mean"])
df_info.apply(f).head(5)
```
